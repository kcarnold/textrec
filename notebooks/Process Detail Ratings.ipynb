```python
import pandas as pd
```

```python
fn = '../data/specificity_details new style completed.xlsx'
```

```python
sheet_names = [name for name in pd.read_excel(fn, sheet_name=None).keys() if 'new' in name]
```

```python
sheets = {}
for name in sheet_names:
    sheets[name] = pd.read_excel(fn, sheet_name=name, header=None)
```

```python
sheets.keys()
```

```python
def fix_index(sheet):
    idx = sheet.iloc[:2].fillna(method='pad', axis=1)
    idx2 = pd.MultiIndex.from_arrays(idx.values)
    names = idx2.to_series().str.join(' ').str.strip()
    names.iloc[0] = 'text'
    sheet = sheet.iloc[2:].copy()
    sheet.columns = names
    return sheet.set_index('text').fillna('x')
```
