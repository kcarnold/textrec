```python
def fix_index(sheet):
    idx = sheet.iloc[:2].fillna(method='pad', axis=1)
    idx2 = pd.MultiIndex.from_arrays(idx.values)
    names = idx2.to_series().str.join(' ').str.strip()
    names.iloc[0] = 'text'
    sheet = sheet.iloc[2:].copy()
    sheet.columns = names
    return sheet.set_index('text')
```
